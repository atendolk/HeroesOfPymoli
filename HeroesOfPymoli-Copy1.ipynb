```python
import pandas as pd 
import json
import numpy as np
```

```python
json_path1 = "./purchase_data.json"
first_data = pd.read_json(json_path1)
first_df = pd.DataFrame(first_data)
first_df
```

```python

players_count = len(first_df["SN"].unique()) 
players_counTable = pd.DataFrame({"Total Players":[players_count]})
players_counTable
```

```python
unique_items = len(first_df["Item ID"].unique())
avg_price=first_df["Price"].mean()
total_purchases=len(first_df["Item ID"])
total_revenue=first_df["Price"].sum()
purchasing_analysis = pd.DataFrame({"Unique Items":[unique_items],
                                   "Avg Price":[avg_price],
                                   "Total Purchases":[total_purchases],
                                   "Total Revenue":[total_revenue]})
purchasing_analysis
```

```python
male_df = first_df.loc[first_df["Gender"] == "Male", :]
female_df = first_df.loc[first_df["Gender"] == "Female", :]
other_df = first_df.loc[first_df["Gender"] == "Other / Non-Disclosed", :]
pmale_df = (len(male_df["SN"])/len(first_df))*100
pfemale_df = (len(female_df["SN"])/len(first_df))*100
pother_df = (len(other_df["SN"])/len(first_df))*100
gender_table = pd.DataFrame({"Gender":["Male", "Female", "Other / Non-Disclosed"],
                                "Percentage of Players":["%.2f" % pmale_df, "%.2f" % pfemale_df, "%.2f" % pother_df],
                                "Total Count":[len(male_df["SN"]), len(female_df["SN"]), len(other_df["SN"])]})
gender_table
```

```python
Themale_df = first_df.loc[first_df["Gender"] == "Male", :]  
Thefemale_df = first_df.loc[first_df["Gender"] == "Female", :]
Theother_df  = first_df.loc[first_df["Gender"] == "Other / Non-Disclosed", :]

Count_male = len(Themale_df["Item ID"])   
Count_female = len(Thefemale_df["Item ID"])
Count_other = len(Theother_df["Item ID"])

Avg_Price_male = Themale_df["Price"].sum()/Count_male
Avg_Price_female = Thefemale_df["Price"].sum()/Count_female
Avg_Price_other = Theother_df["Price"].sum()/Count_other

Tot_Value_male = Themale_df["Price"].sum()
Tot_Value_female = Thefemale_df["Price"].sum()
Tot_Value_other = Theother_df["Price"].sum()

Normalized_Totals_male = Themale_df["Price"].mean()
Normalized_Totals_female = Thefemale_df["Price"].mean()
Normalized_Totals_other = Theother_df["Price"].mean()

gender_purchase_table = pd.DataFrame({" ":["Gender", "Female", "Male", "Other / Non-Disclosed"],
                                "Purchase Count":["", Count_female, Count_male, Count_other],
                                "Average Purchase Price":["", "$%.2f" % Avg_Price_female, "$%.2f" % Avg_Price_male, "$%.2f" % Avg_Price_other],
                                "Total Purchase Value":["" , "$%.2f" % Tot_Value_female, "$%.2f" % Tot_Value_male, "$%.2f" % Tot_Value_other],      
                                "Normalized Totals":["" , "$%.2f" % Normalized_Totals_female, "$%.2f" % Normalized_Totals_male, "$%.2f" % Normalized_Totals_other]})
gender_purchase_table.set_index(" ", inplace=True)
gender_purchase_table
```

```python
bins = [0,9,14,19,24,29,34,39,150] 

labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

reduced_df = first_df[["Age", "Price"]]
reduced_df["Age Group"] = pd.cut(reduced_df["Age"],bins,labels=labels)
Age_Group = reduced_df.groupby("Age Group")

Purchase_Count = Age_Group.size().to_frame(name= "Purchase_Count")          
Average_Purchase_Price = Age_Group["Price"].mean().to_frame(name= "Average_Purchase_Price")   
Total_Purchase_Value = Age_Group["Price"].sum().to_frame(name= "Total_Purchase_Value")     
Normalized_Totals = Age_Group["Price"].sum().to_frame(name= "Normalized_Totals")

Age_Demo_df = pd.concat([Purchase_Count, Average_Purchase_Price, Total_Purchase_Value, Normalized_Totals], axis=1)
Age_Demo_df['Average_Purchase_Price'] = Age_Demo_df['Average_Purchase_Price'].map("${:.2f}".format)
Age_Demo_df['Total_Purchase_Value'] = Age_Demo_df['Total_Purchase_Value'].map("${:.2f}".format)
Age_Demo_df['Normalized_Totals'] = Age_Demo_df['Normalized_Totals'].map("${:.2f}".format)
Age_Demo_df
```

```python
SN_Grouping = first_df.groupby("SN")

Purchase_Count = SN_Grouping.size().to_frame(name= "Purchase_Count")           
Average_Purchase_Price = SN_Grouping["Price"].mean().to_frame(name= "Average_Purchase_Price")   
Total_Purchase_Value = SN_Grouping["Price"].sum().to_frame(name= "Total_Purchase_Value")     
SN_df = pd.concat([Purchase_Count, Average_Purchase_Price, Total_Purchase_Value], axis=1)
SN_df['Average_Purchase_Price'] = SN_df['Average_Purchase_Price'].map("${:.2f}".format)
SN_df['Total_Purchase_Value'] = SN_df['Total_Purchase_Value'].map("${:.2f}".format)
SN1_df = pd.DataFrame(SN_df)
SN1_df = SN1_df.sort_values("Total_Purchase_Value", ascending=False)

SN1_df.head()
```

```python
item = first_df.groupby("Item ID")

Purchase_Count = item["Item ID"].size().to_frame(name= "Purchase Count")  
Price = item["Price"].mean().to_frame(name= "Price")  
Name = item["Item Name"].unique().to_frame(name= "Item Name") 
PValue = item["Price"].sum().to_frame(name= "Total_Purchase_Value")
Igroup_df = pd.concat([Name, Purchase_Count, Price, PValue], axis=1)
Igroup_df['Total_Purchase_Value'] = Igroup_df['Total_Purchase_Value'].map("${:.2f}".format)
Igroup_df['Price'] = Igroup_df['Price'].map("${:.2f}".format)
item_group_df = pd.DataFrame(Igroup_df)
item_group_df = item_group_df.sort_values("Purchase Count", ascending=False)


item_group_df.head() 
```
